# BULDING A LINEAR REGRESSION MODEL 

```python
x = data["PetalLengthCm"]
```

```python
y= data["PetalWidthCm"]
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_train_sm = sm.add_constant(X_train)
```

```python
lr = sm.OLS(Y_train, X_train_sm).fit()
```

```python
lr.params
```

```python
print(lr.summary())
```

```python
print(X_train.head())
print(X_train_sm.head(100))
```

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train, -0.352933 + 0.411117 * X_train, 'r')
plt.show()
```

```python
# we need to check if the error terms are also normally distributed
y_train_pred = lr.predict(X_train_sm)
res = (Y_train - y_train_pred)

fig = plt.figure()
sns.distplot(res)
fig.subplot('Error Terms', fontsize = 15)
plt.xlabel('Y_train - y_train_pred', font_size = 15)
plt.show()
```

```python
#looking for pattern in the residuals
plt.scatter(X_train, res)
plt.show()
```

```python
# Predictions on the test set
X_test_sm = sm.add_constant(x_test)
y_pred = lr.predict(X_test_sm)
```

```python
# looking at the RMSE

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# it has got low rmse value - which is good !
```

```python
r_squared = r2_score(y_test, y_pred)
r_squared
```

```python
r_squared = r2_score(Y_train, y_train_pred)
r_squared
```

```python
# We dont see much difference in data - so the model fits well!!!!
```
