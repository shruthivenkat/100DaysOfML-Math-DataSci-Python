## Analysis 1 : Most common facility type in each district

```python
temp_df = pd.DataFrame(facility.groupby(['District', 'Type']).size().unstack())
temp_df
```

```python
# to get theindex of first occurrence of maximum over requested axis
temp_df = temp_df.idxmax(axis=1)
```

```python
temp_df.head()
```

## Analysis 2 : Most common facility type in each town

```python
temp_df = pd.DataFrame(facility.groupby(["Town", "Type"]).size().unstack().idxmax(axis =1))
temp_df
```

## Analysis 3 : Most common facility type in each region

```python
df_temp = pd.DataFrame(facility.groupby(["Region","Type"]).size().unstack().idxmax(axis = 1))
df_temp
```

## Analysis 4 : Most common type of facility over the country

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
facility_in_numeric = facility['Type'].value_counts()
plt.figure(figsize = (20, 12))
sns.barplot(x = facility_in_numeric.index, y = facility_in_numeric.values)
plt.xticks(rotation = 90)
plt.show()
```

## Analysis 5 : Who owns the maximum number of facilities?

```python
ownership_counts = facility['Ownership'].str.capitalize().value_counts()
ownership_counts
```
