## Here, we can observe that there are 10052 rows with null value. so, its better not to consider the columns 'last_review' & 'reviews_per_month'.

```python
data = data.dropna(subset=['last_review','reviews_per_month'], axis = 0)
data.isna().sum()
```

```python
data.describe()
```

```python
data
```

### well, to make an analysis for the rooms in Air BnB, we just need their id or their names. So, lets drop one of these columns

```python
data = data.drop("name", axis = 1)
data.head()
```

### similarly, we just need one of the columns to define the host - either the name or his/her host_id. Lets keep the host_id and drop the host name. (there are 16 rows with no host_name. so lets decide to take the host id)

```python
data = data.drop("host_name", axis = 1)
data.head()
```

### Do we have anything to do with the lat and long of room location??? hmm.. well, i dont think so.

### We have the neighbourhood column to talk about the geographyical location. we may not find any interesting facts by using the exact lat and long. So, lets just drop them!

```python
data = data.drop(["latitude","longitude"], axis=1)
data.head()
```

## Data Visuvalisation and Inference

```python
data["neighbourhood_group"].unique()
```

```python
plt.figure(figsize=(15,8))
sns.set(style="whitegrid")
plt.title('Count of rooms in each neighbourhood', size=30)
plt.xlabel("Neighbourhood", size = 25)
plt.ylabel("count",size=25)
sns.countplot(data.neighbourhood_group,palette="plasma")
plt.show()
```

## ANALYSIS : neighbourhood in NYC with large number of Air BnB rooms = "Manhattan"

```python
plt.figure(figsize=(15,8))

sns.countplot(data.room_type,palette='rocket')
plt.title('Count of each room type in NYC', size = 30)
plt.xlabel('Room type', size=25)
plt.ylabel('Count', size = 25)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data.sort_values('neighbourhood_group').neighbourhood_group,hue=data.room_type, palette='plasma')
plt.title('Neighbourhood wise room type count', size = 25)
plt.xlabel('Neighbourhood name', size = 25)
plt.ylabel('Count', size = 25)
plt.show()
```

## It can be seen that most of the bookings happen in Manhattan than any other Neighbourhoods in NYC

Well, lets see why this is happening in depth

```python
plt.yscale('log')
sns.boxplot(x="neighbourhood_group", y="price", data=data[data.room_type=='Shared room'], palette='plasma')
plt.title("Shared room",size=30)
```

### Observations - "Shared rooms" :

Air BNB's shared rooms in 'Staten island' is said to have a comparitively wider range of price for their rooms. Starting from a really lower end to a fairly higher end.

Rooms in 'Bronx' starts from the mid range of rooms in 'Staten island'.


Most of the rooms in 'Brooklyn' is said to have mid range tarrifs. But, if you'd see the outliers, it gives us an additional info that there are a small number of super costly rooms. 

Manhatten seems to be a perfect place for any kind of people to live in. 

if it comes to budgeted shared rooms, manhatten has the lowest priced rooms. but, moving into these cheapest shared rooms must be a big deal cos of the number of available rooms. most of the rooms in manhatten is said to be really expensive than other parts of the city. 

Probably that is the reason why there are a very less number of shared rooms in the cheapest rate of the city. (to make living affordable to atleast few budgeted individuals)

```python
neigh = data.neighbourhood_group.unique()
for i in range(0,len(neigh)):
    data_neigh = data[data.neighbourhood_group == neigh[i]]
    print(data_neigh.neighbourhood.unique())
    data_neigh.neighbourhood.value_counts().sort_values(ascending=False)[:10].sort_values().plot(kind='barh')
    plt.title("Top 10 neighbourhood in "+ neigh[i] +" with high number of Air BnB rentals",size=20)
    plt.show()
    
```
